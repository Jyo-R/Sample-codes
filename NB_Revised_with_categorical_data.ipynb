```python
#################################################################################
#################################################################################
########################### AIML Mid Term Assignment  ###########################
########################### Naive Bayes Algorithm     ###########################
########################### Columns-2,3,4,5 + Outcome ###########################
########################### By:Jyothiswaroop Rajan.   ###########################
########################### Roll # - XP22012          ###########################
#################################################################################
#################################################################################

#Required imports
import numpy as np
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay

#Font formatting codes for beautifying the output to a more readable form - in print statements
BOLD = '\033[1m' #Plain - bold
BL =  '\033[36m' #Blue font
BB = BL + BOLD #Blue and Bold - to highlight
UL = '\033[4m'
BULL = UL + '\033[96m' + BOLD #Underline + Bold (in blue)- for headings
END = '\033[0m' #End format

def printhead(arg):
  print(BULL, end ="") 
  print(arg)
  print("\n")
  print(END, end ="")

def printhigh(arg):
  print(BB, end ="")  
  print(arg)
  print("\n")
  print(END, end ="")

def printUL(arg):
  print(UL, end ="") 
  print(arg)
  print("\n")
  print(END, end ="")

def printb(arg):
  print(BL, end ="") 
  print(arg)
  print("\n")
  print(END, end ="")


#Read the data from the CSV
data = pd.read_csv('/content/Jyo_DiabetesDataset.csv')

#Setting column display limit for tables
pd.set_option('display.expand_frame_repr', False) 

#Display Original Data and its size
printhead("Original Data Given:")
printUL("Size of the given data:")
printhigh(data.shape)

printUL("Records Sample:")
printb(data.head())


#Clean up the data
printhead("Performing Data Clean Up:");
printUL("Considering only given fields (2,3,4,5,Outcome)")

#Removing fields that are not required
data.drop('Pregnancies', inplace=True, axis=1)
data.drop('DiabetesPedigreeFunction', inplace=True, axis=1)
data.drop('Age', inplace=True, axis=1)
data.drop('BMI', inplace=True, axis=1)

#Removing outliers and 0 records that may skew the outcome
printUL("Removing records that contain 0s and outliers")
df = pd.DataFrame(np.random.randn(100, 3))
df[(np.abs(stats.zscore(df)) < 3).all(axis=1)]
data = data[data.Glucose != 0]
data = data[data.BloodPressure != 0]
data = data[data.SkinThickness != 0]
data = data[data.Insulin != 0]

printhead("Data clean up completed - Data after clean up:")
printb(data.head())
printb(data.shape)

#Visualize the cleaned up data for our fields alone as a histogram
printhead("A simple box/barplot to show the prior probability for each selected classes:")
plt.figure(figsize=(12,12))
plt.subplot(3,3,1)
sns.barplot(x='Outcome',y='Glucose',data=data,hue="Outcome")
plt.subplot(3,3,2)
sns.barplot(x='Outcome',y='BloodPressure',data=data,hue="Outcome")
plt.subplot(3,3,3)
sns.barplot(x='Outcome',y='SkinThickness',data=data,hue="Outcome")
plt.subplot(3,3,4)
sns.barplot(x='Outcome',y='Insulin',data=data,hue="Outcome")
plt.show()
print("\n")

#A pair plot - to visualize and depict the relation to diabetes and to gauge the most seperated clusters
printhead("A simple pairplot showing the relation to outcome:")
aa=sns.pairplot(data,hue='Outcome')
plt.show()

printhead("Performing data discretization")
#Performing data discretization - for converting numerical to nominal for evaluating NB

#labels for discretization - We will re-label the numerical to nominal with these range labels
labels = ['Low','Medium','High']

#Preprocessing
for j in data.columns[:-1]:
    mean = data[j].mean()
    data[j] = data[j].replace(0,mean)
    data[j] = pd.cut(data[j],bins=len(labels),labels=labels)

printUL("Discretized data (Final data set for evaluating the Naive Bayes Model):")
printb(data.head())

#Training split - 80 / 70/ 60 per iteration

split_per = [80,70,60]

def count(data,colname,label,target):
    condition = (data[colname] == label) & (data['Outcome'] == target)
    return len(data[condition])


for i in split_per:

    #Split training and testing data based on the split; 1st iteration (80-20)- 2nd (70-30) - 3rd (60-40)
    printhead("Iteration: " + str(i))
    predicted = []
    probabilities = {0:{},1:{}}
    train_len = int((i*len(data))/100)
    printhead("Running for training length = " + str(train_len))

    #Consider ALL columns from the final input data set
    trainData = data.iloc[:train_len,:]
    
    #Strip off Outcome field - use as base test data
    testData = data.iloc[train_len+1:,:-1] 

    #Output field for the test data - To compare actual outcome against predicted
    actual = data.iloc[train_len+1:,-1]

    count_0 = count(trainData,'Outcome',0,0)
    count_1 = count(trainData,'Outcome',1,1)
    
    ####Calculating Probabilities and Likelihoods######
    
    #Probabilities of outcome - 0's - non diabetic and 1's - diabetic
    prob_0 = count_0/len(trainData)
    prob_1 = count_1/len(trainData)
    printhigh("P(X) = "+str(prob_0))
    
    #Train the model for the split taken
    printhead("Training the model with the obtained probabilities")
    for j in trainData.columns[:-1]:
        probabilities[0][j] = {}
        probabilities[1][j] = {}
        for k in labels:
            count_k_0 = count(trainData,j,k,0)
            count_k_1 = count(trainData,j,k,1)
            probabilities[0][j][k] = count_k_0 / count_0
            probabilities[1][j][k] = count_k_1 / count_1

    #Testing the model for the obtained split
    printhead("Testing the model - test set - Length:" + str(len(testData)))
    for row in range(0,len(testData)):
        prod_0 = prob_0
        prod_1 = prob_1
        for feature in testData.columns:
            prod_0 *= probabilities[0][feature][testData[feature].iloc[row]]
            prod_1 *= probabilities[1][feature][testData[feature].iloc[row]]
        
    #Result for test run - Predicted values
        if prod_0 > prod_1:
            predicted.append(0)
        else:
            predicted.append(1)
    
    printhigh("Predicted Outcome values for each row (as an array)-" +str(predicted))

    #Confusion Matrix for prediction vs actual
    tp,tn,fp,fn = 0,0,0,0
    for j in range(0,len(predicted)):
        if predicted[j] == 0:
            if actual.iloc[j] == 0:
                tp += 1
            else:
                fp += 1
        else:
            if actual.iloc[j] == 1:
                tn += 1
            else:
                fn += 1

    printhigh("TruePositive = "+str(tp))
    printhigh("TrueNegative = "+str(tn))
    printhigh("FalsePositive = "+str(fp))
    printhigh("FalseNegative = "+str(fn))

    cm=confusion_matrix(np.array(actual),np.array(predicted))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()
    plt.show()


    printhigh('Accuracy for training length '+str(i)+'% : '+str(((tp+tn)/len(actual))*100))
```
